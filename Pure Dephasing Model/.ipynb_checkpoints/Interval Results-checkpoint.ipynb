```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Configure font properties for the plots
plt.rc('font',family='Times New Roman')
plt.rcParams.update({'font.size':13})
plt.rcParams['font.size'] = 13
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'
plt.rcParams['axes.linewidth'] = 1
```

```python
#loading the loss values for each interval length
trainingloss = np.loadtxt('traininglossvintervallength.csv', delimiter=',')
testloss = np.loadtxt('testlossvintervallength.csv', delimiter=',')

#Define an array of interval lengths, starting from 0.25 and increasing by 0.2 up to 2.05. Subtract 0.25 to adjust 
#the range so that it represents the actual interval length.
intervallength = np.arange(0.25, 2.05+0.2, 0.2) - 0.25
```

```python
#Plotting the training and test loss against the interval length
plt.plot(intervallength , trainingloss, '--', color='blue', label='Training MSE')
plt.plot(intervallength, testloss, color='green', label='Test MSE')
plt.xlabel('Interval Length')
plt.ylabel('MSE')
plt.legend()
plt.savefig('lossvintervallength.pdf', bbox_inches='tight')
```

```python
#Set the upper bound for loading the dataset with the shortest non-zero interval length
upper_bound = 0.45

#Load the training labels from the csv file
Y = np.loadtxt('Data/Ytrain_varyηandω_0.25and{0}.csv'.format(np.round(upper_bound, 2)), delimiter=',')

#Define the number of samples for the training, validation and test sets
ntrain = 4800
nval = 2400
ntest = 2400

#Extract the test set labels
Ytest = Y[ntrain+nval:ntrain+nval+ntest, 3:6]

#Load the predictions from the test set
predictionstest = np.loadtxt('Predictions/predictionstest_0.25and{0}.csv'.format(np.round(upper_bound, 2)), delimiter=',')
```

```python
#Create a scatter plot of predicted values for $\omega_c$ versus true values
plt.plot(Ytest[:,0], predictionstest[:,0],'bo', label='Predictions')
plt.plot(Ytest[:,0], Ytest[:,0], color='orange', label='True')
plt.ylabel(r"$\mathrm{\hat{\omega}_c / \omega_0}$")
plt.xlabel(r"$\mathrm{\omega_c / \omega_0}$")
plt.legend()
plt.savefig('predictedomegacvreal_0.25and0.45.pdf', bbox_inches='tight')
plt.show()
```

```python
#define an array where each element is the difference between the predicted value and true value for $\omega_c$ for
#a given trajectory
diffsomegac0p45 = predictionstest[:,0] - Ytest[:,0]

#print the minimum difference to find the smallest prediction error
print(np.amin(diffsomegac0p45))
#print the maximum difference to find the largest prediction error
print(np.amax(diffsomegac0p45))
```

```python
#define intervals for analysing the differences between predicted and true values for $|omega_c$
intervalsomegac0p45 = [-0.016, -0.0155, -0.015, -0.0145, -0.014, -0.0135, -0.013, -0.0125, -0.012, -0.0115, -0.011, -0.0105, -0.010, -0.0095, -0.009, -0.0085, -0.008, -0.0075, -0.007, -0.0065, -0.006, -0.0055, -0.005, -0.0045, -0.004, -0.0035, -0.003, -0.0025, -0.002, -0.0015, -0.001,-0.0005, 0.0, 0.0005, 0.001, 0.0015, 0.002, 0.0025, 0.003, 0.0035, 0.004, 0.0045, 0.005, 0.0055, 0.006]
```

```python
#initialise list to store the frequencies of differences in intervals for $|omega_c$
freqomegac0p45 = []

#loop through each interval
for i in range(len(intervalsomegac0p45)-1):
    #create a mask to find differences within the current interval
    maskomegac0p45 = (diffsomegac0p45 >= intervalsomegac0p45[i]) & (diffsomegac0p45 < intervalsomegac0p45[i+1])
    #append the count of differences within the current interval to the frequency list
    freqomegac0p45.append(len(diffsomegac0p45[maskomegac0p45]))
```

```python
#calculate x positions for bars and heights
xomegac0p45 = intervalsomegac0p45[1:] 
heightsomegac0p45 = (np.array(freqomegac0p45)/2400)*100

#create the bar chart
plt.bar(xomegac0p45, heightsomegac0p45, width=-0.0005, align='edge', alpha = 1, color = 'orange', edgecolor='black')  
plt.xlabel(r"$\mathrm{(\hat{\omega}_c - \omega_c)/\omega_0}$")
plt.ylabel('% of trajectories')
plt.savefig('omegacbarchart0.45.pdf', bbox_inches='tight')
plt.show()
```
