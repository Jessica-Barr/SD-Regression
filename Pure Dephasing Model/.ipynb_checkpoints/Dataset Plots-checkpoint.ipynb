```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
#Loading the data with $\eta = 0.25$, $\omega_c = 0.5$ with $s \in (0, 0.5]$ if the spectral density is sub-Ohmic,
#$s \in [1.5, 4]$ is the spectral density is super-Ohmic and $s=1$ is the spectral density is Ohmic.
Xtrain_separated = np.loadtxt('Data/Xtrainx_fixedηandω_separated.csv', delimiter=',')
Ytrain_separated = np.loadtxt('Data/Ytrain_fixedηandω_separated.csv', delimiter=',')

#Loading the data with $\eta = 0.25$, $\omega_c = 0.5$ with $s \in (0, 1)$ if the spectral density is sub-Ohmic,
#$s \in (1, 4]$ is the spectral density is super-Ohmic and $s=1$ is the spectral density is Ohmic.
Xtrain_notseparated = np.loadtxt('Data/Xtrainx_fixedηandω_notseparated.csv', delimiter=',')
Ytrain_notseparated = np.loadtxt('Data/Ytrain_fixedηandω_notseparated.csv', delimiter=',')

#Loading the data with $\eta = 0.25$, $\omega_c = 0.25$ with $s \in (0, 1)$ if the spectral density is sub-Ohmic,
#$s \in (1, 4]$ is the spectral density is super-Ohmic and $s=1$ is the spectral density is Ohmic.
Xtrain_0p250p25 = np.loadtxt('Data/Xtrainx_varyηandω_0.25and0.25.csv', delimiter=',')
Ytrain_0p250p25 = np.loadtxt('Data/Ytrain_varyηandω_0.25and0.25.csv', delimiter=',')

#Loading the data with $\eta, \omega_c \in [0.25, 2.05]$, with $s \in (0, 1)$ if the spectral density is sub-Ohmic,
#$s \in (1, 4]$ is the spectral density is super-Ohmic and $s=1$ is the spectral density is Ohmic.
Xtrain_0p252p05 = np.loadtxt('Data/Xtrainx_varyηandω_0.25and2.05.csv', delimiter=',')
Ytrain_0p252p05 = np.loadtxt('Data/Ytrain_varyηandω_0.25and2.05.csv', delimiter=',')
```

```python
#Configure font properties for the plots
plt.rc('font',family='Times New Roman')
plt.rcParams.update({'font.size':13})
plt.rcParams['font.size'] = 13
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'
plt.rcParams['axes.linewidth'] = 1
```

```python
#Define start and end points of the time interval
a = 0 #start time
b = 10 #end time

#Set the time step size
dt = 0.025

#Create an array of time points from a to b with increments of dt
t = np.arange(a, b+dt, dt)
```

```python
#Import the 'viridis' colormap from matplotlib
cmap = cm.viridis

#Function to map a value to the [0,1] range for use with a colormap
def map_to_colormap_range(value, min_val, max_val):
    #normalise value to a range between 0 and 1 based on min_val and max_val
    return(value-min_val)/(max_val-min_val)
```

```python
#Ytrain_separated[:,[4]] selects the column with the Ohmicity parameter, located at index 4
svals_separated = Ytrain_separated[:,4]

#Loop over the first 100 training samples
for i in range(Xtrain_separated[0:100,:].shape[0]):
    #Map the ohmicity parameter to a range suitable for the colormap
    color_value = map_to_colormap_range(svals_separated[i], np.amin(svals_separated), np.amax(svals_separated))
    #Plot the time evolution of the first 100 samples, with colour determined by the Ohmicity parameter
    plt.plot(t, Xtrain_separated[i,:], color=cmap(color_value), linewidth = 0.1)

#Label the axis
plt.xlabel(r"$ \mathrm{\omega_0 t}$ ")
plt.ylabel(r"$\mathrm{ \langle \sigma_x (t) \rangle }$")
#Save the plot as a pdf file
plt.savefig('trainingset_fixedetaomegac_separated_regression.pdf', bbox_inches='tight')
#Display the plot
plt.show()
```

```python
#Ytrain_notseparated[:,4] selects the column with the Ohmicity parameter, located at index 4
svals_notseparated = Ytrain_notseparated[:,4]

#Loop over the first 100 training samples
for i in range(Xtrain_notseparated[0:100,:].shape[0]):
    #Map the Ohmicity parameter to a suitable range for the colormap
    color_value = map_to_colormap_range(svals_notseparated[i], np.amin(svals_notseparated), np.amax(svals_notseparated))
    #Plot the time evolution of the first 100 samples, with colour determined by the Ohmicity parameter
    plt.plot(t, Xtrain_notseparated[i,:], color=cmap(color_value), linewidth = 0.1)

#Label the axis
plt.xlabel(r"$ \mathrm{\omega_0 t}$ ")
plt.ylabel(r"$\mathrm{ \langle \sigma_x (t) \rangle }$")
#Save the plot as a pdf file
plt.savefig('trainingset_fixedetaomegac_notseparated_regression.pdf', bbox_inches='tight')
#Display the plot
plt.show()
```

```python
#Ytrain_0p250p25[:,4] selects the column with the Ohmicity parameter, located at index 4
svals_0p250p25 = Ytrain_0p250p25[:,4]

#Loop over the first 100 training samples
for i in range(Xtrain_0p250p25[0:100,:].shape[0]):
    #Map the Ohmicity parameter to a suitable range for the colormap
    color_value = map_to_colormap_range(svals_0p250p25[i], np.amin(svals_0p250p25), np.amax(svals_0p250p25))
    #Plot the time evolution of the first 100 samples, with colour determined by the Ohmicity parameter
    plt.plot(t, Xtrain_0p250p25[i,:], color=cmap(color_value), linewidth = 0.1)

#Label the axis
plt.xlabel(r"$ \mathrm{\omega_0 t}$ ")
plt.ylabel(r"$\mathrm{ \langle \sigma_x (t) \rangle }$")
#Save the plot as a pdf file
plt.savefig('trainingset_varyetaomegac_0.25and0.25_regression.pdf', bbox_inches='tight')
#Display the plot
plt.show()
```

```python
#Ytrain_0p252p05[:,4] selects the column with the Ohmicity parameter, located at index 4
svals_0p252p05 = Ytrain_0p252p05[:,4]

#Loop over the first 100 training samples
for i in range(Xtrain_0p252p05[0:100,:].shape[0]):
    #Map the Ohmicity parameter to a suitable range for the colormap
    color_value = map_to_colormap_range(svals_0p252p05[i], np.amin(svals_0p252p05), np.amax(svals_0p252p05))
    #Plot the time evolution of the first 100 samples, with colour determined by the Ohmicity parameter
    plt.plot(t, Xtrain_0p252p05[i,:], color=cmap(color_value), linewidth = 0.1)
    
#Label the axis
plt.xlabel(r"$ \mathrm{\omega_0 t}$ ")
plt.ylabel(r"$\mathrm{ \langle \sigma_x (t) \rangle }$")
#Save the plot as a pdf file
plt.savefig('trainingset_varyetaomegac_0.25and2.05_regression.pdf', bbox_inches='tight')
#Display the plot
plt.show()
```

```python
#Create a 2x2 grid of subplots with a specified figure size
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18,10))

#Plot the first 200 trajectories from the 'separated' training dataset on the first subplot
for i in range(Xtrain_separated[0:200,:].shape[0]):
    #Map the Ohmicity parameter value to a colormap range
    color_value = map_to_colormap_range(Ytrain_separated[i,4], np.amin(Ytrain_separated[:,4]), np.amax(Ytrain_separated[:,4]))
    #Plot the trajectory on the top-left subplot
    axes[0,0].plot(t, Xtrain_separated[i,:], color=cmap(color_value), linewidth = 0.1)
    
#Plot the first 200 trajectories from the 'not separated' training dataset on the second subplot
for j in range(Xtrain_notseparated[0:200,:].shape[0]):
    color_value = map_to_colormap_range(Ytrain_notseparated[j,4], np.amin(Ytrain_notseparated[:,4]), np.amax(Ytrain_notseparated[:,4]))
    axes[0,1].plot(t, Xtrain_notseparated[j,:], color=cmap(color_value), linewidth = 0.1)

#Plot the first 200 trajectories from the dataset where $\eta = \omega_c = 0.25$ on the third subplot
for k in range(Xtrain_0p250p25[0:200,:].shape[0]):
    color_value = map_to_colormap_range(Ytrain_0p250p25[k,4], np.amin(Ytrain_0p250p25[:,4]), np.amax(Ytrain_0p250p25[:,4]))
    axes[1,0].plot(t, Xtrain_0p250p25[k,:], color=cmap(color_value), linewidth = 0.1)

#Plot the first 200 trajectories from the dataset where $\eta, \omega_c = 2.05$ on the fourth subplot
for l in range(Xtrain_0p252p05[0:200,:].shape[0]):
    color_value = map_to_colormap_range(Ytrain_0p252p05[l,4], np.amin(Ytrain_0p252p05[:,4]), np.amax(Ytrain_0p252p05[:,4]))
    axes[1,1].plot(t, Xtrain_0p252p05[l,:], color=cmap(color_value), linewidth = 0.1)
    
#Set the y-axis labels for all subplots
axes[0,0].set_ylabel(r"$\mathrm{\langle \sigma_x (t) \rangle}$")
axes[0,1].set_ylabel(r"$\mathrm{\langle \sigma_x (t) \rangle}$")
axes[1,0].set_ylabel(r"$\mathrm{\langle \sigma_x (t) \rangle}$")
axes[1,1].set_ylabel(r"$\mathrm{\langle \sigma_x (t) \rangle}$")

#Set the x-axis labels for all subplots
axes[0,0].set_xlabel(r"$\mathrm{\omega_0 t }$")
axes[0,1].set_xlabel(r"$\mathrm{\omega_0 t }$")
axes[1,0].set_xlabel(r"$\mathrm{\omega_0 t }$")
axes[1,1].set_xlabel(r"$\mathrm{\omega_0 t }$")

#Create a colorbar for the colormap based on the Ohmicity parameter
sm = plt.cm.ScalarMappable(cmap=cmap)
sm.set_array(svals_separated)
cbar = fig.colorbar(sm, ax=axes.ravel().tolist(), aspect=40)
cbar.set_label(r"$s$", rotation=270, labelpad=15)

#Add panel labels (a), (b), (c), (d) to each subplot
panel_labels = ['(a)', '(b)', '(c)', '(d)']
for i, ax in enumerate(axes.flat):
    ax.text(-0.1, 1.1, panel_labels[i], transform=ax.transAxes, fontsize=14, va='top', ha='right')

#Save the figure as a pdf
plt.savefig('trainingsets_regression.pdf', bbox_inches='tight')
```
